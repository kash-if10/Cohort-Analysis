# Importing Libraries

```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing Dataset and Gathering informatioon regarding Dataset

```python
retail = pd.read_excel('online retail.xlsx')
retail.head(5)
```

```python
retail.info()
```

# Preprocessing the DATA

## Cleaning Data

### Drop Rows with no customer ID

```python
retail = retail.dropna(subset=['CustomerID'])
retail.info()
```

### Create an Invoice Month

```python
#Returning the 1st day of the month 

def get_month(x):
    return dt.datetime(x.year, x.month,1)    
```

```python
# Applying the Function
retail['InvoiceMonth'] = retail['InvoiceDate'].apply(get_month)
```

```python
retail.head()
```

### Create a Cohort Month

### create a column when the  first time customer was acquired

```python
retail['Cohort Month'] =  retail.groupby('CustomerID')['InvoiceMonth'].transform('min')
retail.tail(10)
```

# Creation of Cohort

## We now have to create a cohort and for that we have to obtain the difference btw when the customer is first acquired and current status or last used. ThereFore Difference btw Cohort Month and Invoice Month

### We Have to Create a Function to calculate The Difference

### Function to Convert Column into day, Month, Year Indivisually

```python
def get_date_elements(df, column):
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year
    return day, month, year 
```

Applying on Invoice Month Column and Cohort Month COlumn

```python
inv_day, inv_month, inv_year = get_date_elements(retail, 'InvoiceMonth')
Cohort_day, Cohort_month, Cohort_year = get_date_elements(retail, 'Cohort Month')
```

Check The Series for one of them

```python
Cohort_year[20:26] # Randomly Showing only 5 of them instead of whole 
```

### Create Cohort Index For Difference

```python
year_diff = inv_year -Cohort_year
month_diff = inv_month - Cohort_month
```

```python
retail['CohortIndex'] = year_diff*12+month_diff+1
retail.tail()
```

0 is the month of acquisition

# Create a Pivot Table

Group The Cohort Month and Cohort Index and count each indivisual Coustomer ID

Cohort Month FOr knowing When they were acquired and Cohort INdex for How long they have been our customer and using our product/Service

```python
# This pd.Series.unique returns the actual Value where as the pd.Series.nunique return the count of those unique values

cohort_data = retail.groupby(['Cohort Month','CohortIndex'])['CustomerID'].apply(pd.Series.unique) 
cohort_data
```

Thiis contains gaps and is not in Tabular Format as due to groupby the result is in multiindex format but we require it in table format

```python
cohort_data = retail.groupby(['Cohort Month','CohortIndex'])['CustomerID'].apply(pd.Series.nunique).reset_index()
cohort_data
```

#### Create A pivot Table

```python
cohort_table = cohort_data.pivot(index='Cohort Month', columns=['CohortIndex'],values='CustomerID')
cohort_table
```

# Visualisation

## Cohort Table in a Heat MAP

```python
plt.figure(figsize=(21,15))
sns.heatmap(cohort_table,annot=True,cmap='Blues')
```

### Fixing THe Index

```python
cohort_table.index = cohort_table.index.strftime('%B %Y') 
```

```python
plt.figure(figsize=(21,15))
sns.heatmap(cohort_table,annot=True,cmap='Blues')
```

## Pivot Table For Percentages

Devide Each Row(axis = 0)   By the Total (First Column)

```python
new_cohort_table = cohort_table.divide(cohort_table.iloc[:,0],axis=0)
new_cohort_table
```

## Visualise The Percentages Pivot Table by Heatmap

```python
plt.figure(figsize=(21,10))
sns.heatmap(new_cohort_table,annot=True,cmap='magma',fmt='.0%') #fmt means format
```
